```python
df.drop('customerID', axis = 1, inplace = True)
```

```python
df.info()
```

```python
X = pd.get_dummies(df.drop('Churn', axis = 1), drop_first = True)
```

```python
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

```python
tree_initial = DecisionTreeClassifier()
```

```python
param_grid = {'criterion' :['gini', 'entropy'],
              'min_samples_split':[2,3],
              'max_depth':[1,3,4,7,9,10,15,20],
              'min_samples_leaf':[1,2], 
              'max_features' : ['auto', 'sqrt', 'log2']}
```

```python
grid_tree = GridSearchCV(tree_initial, param_grid, cv = 5)
```

```python
grid_tree.fit(X_train,y_train)
```

```python
grid_tree_pred = grid_tree.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_estimator(grid_tree, X_test,y_test);
```

```python
print(classification_report(y_test, grid_tree_pred))
```

```python
grid_tree.best_estimator_
```

```python
grid_tree.best_params_
```

```python
feat_imp = grid_tree.best_estimator_.feature_importances_
```

```python
feat_imp = pd.DataFrame(data = feat_imp, index = X.columns)
```

```python
feat_imp.columns = ['importance']
```

```python
feat_imp.sort_values('importance', inplace = True)
```

```python
sns.barplot(data = feat_imp, x = feat_imp.index, y = 'importance')
plt.xticks(rotation = 90);
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(grid_tree.best_estimator_,filled=True,feature_names=X.columns);
```

```python
ada_boost = AdaBoostClassifier()
```

```python
param_grid = {'n_estimators': [40,50,60,70],
              'learning_rate': [0.5,1.0,1.5,2.0]}
```

```python
grid_ada = GridSearchCV(ada_boost, param_grid, cv = 5)
```

```python
grid_ada.fit(X_train,y_train)
```

```python
grid_ada_pred = grid_ada.predict(X_test)
```

```python
ConfusionMatrixDisplay.from_estimator(grid_ada, X_test,y_test);
```
